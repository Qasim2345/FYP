```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for column in data.columns:
    data[column]= le.fit_transform(data[column])
    #change any variable with corresponding integer value
```

```python
#now split data into train and test
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = data.drop('class', axis=1)
y = data['class']
#These lines split the dataset into feature matrix X (i.e., all columns except the 'class' column) 
#and target vector y (i.e., the 'class' column).
```

```python
from sklearn.decomposition import PCA
pca1 = PCA(n_components = 11)
pca_fit1 = pca1.fit_transform(X)
#PCA is a technique used for feature reduction or dimensionality reduction, 
#which aims to reduce the number of features in the dataset while retaining as much information as possible.
```

```python
X_train, X_test,y_train,y_test = train_test_split(pca_fit1,y,test_size = 0.3,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier #import models
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
#train models on dataset
knn=KNeighborsClassifier()
knn.fit(X_train, y_train)

svc = SVC()
svc.fit(X_train,y_train)

dt = DecisionTreeClassifier()
dt.fit(X_train,y_train)

lr = LogisticRegression()
lr.fit(X_train,y_train)

rm = RandomForestClassifier()
rm.fit(X_train, y_train)

gb = GradientBoostingClassifier()
gb.fit(X_train, y_train)

gnb = GaussianNB()
gnb.fit(X_train, y_train)
```

```python
#predict result
y_pred1 = knn.predict(X_test)
y_pred2 = svc.predict(X_test)
y_pred3 = dt.predict(X_test)
y_pred4 = lr.predict(X_test)
y_pred5 = rm.predict(X_test)
y_pred6 = gb.predict(X_test)
y_pred7 = gnb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
#check accuracy of models
print("ACC KNN", accuracy_score(y_test,y_pred1))
print("ACC SVC", accuracy_score(y_test,y_pred2))
print("ACC DT", accuracy_score(y_test,y_pred3))
print("ACC LR", accuracy_score(y_test,y_pred4))
print("ACC RM", accuracy_score(y_test,y_pred5))
print("ACC GBC", accuracy_score(y_test,y_pred6))
print("ACC GNB", accuracy_score(y_test,y_pred7))
```

```python
final_data = pd.DataFrame({'Models':['KNN','SVC','DT','LR','RM','GBC','GNB'],
                          'ACC': [accuracy_score(y_test,y_pred1)*100,
                                 accuracy_score(y_test,y_pred2)*100,
                                 accuracy_score(y_test,y_pred3)*100,
                                 accuracy_score(y_test,y_pred4)*100,
                                 accuracy_score(y_test,y_pred5)*100,
                                 accuracy_score(y_test,y_pred6)*100,
                                 accuracy_score(y_test,y_pred7)*100]})
```
